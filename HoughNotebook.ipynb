```python
#rng,img_gray = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(img_gray, cmap='gray')
```

```python
circles = cv2.HoughCircles(img_gray, cv2.cv.CV_HOUGH_GRADIENT,
          1.0,
          0.001,
          param1=100,
          param2=3.5,
          minRadius=2,
          maxRadius=4)[0]

result = []

for c in circles:
    x, y, radius = c
    center = (int(x), int(y))
    radius = int(radius)
    # if radius < 0 or radius > 50:
    #     continue
    #if draw:
    cv2.circle(img_gray, center, radius, (0, 255, 0), 1)
    result.append([center[0], center[1], radius])
```
