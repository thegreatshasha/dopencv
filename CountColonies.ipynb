Ok this size is fine for applying algos> Let's try out different thresholding methods now

In order to compare it to adaptive, otsu's method and canary, let's also plot these as well.

```python
rect,thresh_otsu = cv2.threshold(img,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(thresh_otsu, cmap='gray')
```

Let's plot the canny as well

```python
from detect import auto_canny
thresh_canny = auto_canny(thresh_otsu)
plt.imshow(thresh_canny, cmap='gray')
```

While we are at it, why not try erosion as well?

```python
circles = cv2.HoughCircles(thresh_canny, cv2.cv.CV_HOUGH_GRADIENT,
              1,
              4,
              200,
              100,
              1, 9)[0]

for circle in circles:
        x, y, radius = circle
        center = (int(x), int(y))
        cv2.circle(img, center, radius, 0, 1)
        
plt.imshow(img, cmap='gray')
```
